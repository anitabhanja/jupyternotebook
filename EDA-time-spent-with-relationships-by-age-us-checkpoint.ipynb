# Who Americans Spend Their Time With?

As humans, we have heavily relied on cooperation and social relationships to thrive. Of course, who we spend time with evolves throughout our lifetime.

Using the American Time Use survey data from "Our World in Data" (https://ourworldindata.org/time-with-others-lifetime), we look at how Americans spend their time with at various ages of their life.

```python
#Time spent with relationships by age in USA
#https://ourworldindata.org/time-with-others-lifetime

#1. Importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
#2. Loading the data into the data frame.
df = pd.read_csv("C:/Users/time-spent-with-relationships-by-age-us.csv")
# To display the top 5 rows 
df.head(5) 
```

```python
#3. Checking the types of data
df.dtypes
```

```python
#4. Dropping irrelevant columns
df = df.drop(['Entity', 'Code'], axis=1)
df.head(5)
```

```python
#5. Renaming the columns
df = df.rename(columns={"Year": "Age",
                        "Time spent alone, by age of respondent (United States)": "Alone", 
                        "Time spent with friends, by age of respondent (United States)": "Friends", 
                        "Time spent with children, by age of respondent (United States)": "Children", 
                        "Time spent with with parents, siblings and other family, by age of respondent (United States)": "Family",
                        "Time spent with partner, by age of respondent (United States)": "Partner", 
                        "Time spent with coworkers, by age of respondent (United States)": "Co-workers"})
df.head(5)
```

```python
#6. Check for duplicate rows and remove if any
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df)
```

```python
#7. Check for missing or null values.
print(df.isnull().sum())
```

There are no duplicate rows, no missing or null values in the data. The data can be considered clean for analysis and visualization.

```python
#8. Plot the line chart

# multiple line plots
plt.figure(figsize=(18,10))
plt.plot(df['Age'], df['Alone'], color='red', label='Alone', linewidth=3.0, marker='o')
plt.plot(df['Age'], df['Friends'], color='blue', label='Friends', linewidth=3.0, marker='o')
plt.plot(df['Age'], df['Children'], color='purple', label='Children', linewidth=3.0, marker='o')
plt.plot(df['Age'], df['Family'], color='green', label='Family', linewidth=3.0, marker='o')
plt.plot(df['Age'], df['Partner'], color='orange', label='Partner', linewidth=3.0, marker='o')
plt.plot(df['Age'], df['Co-workers'], color='yellow', label='Co-workers', linewidth=3.0, marker='o')

# Add title and axis names
plt.title("Who Americans Spend Their Time With?")
plt.xlabel('Age of person')
plt.ylabel('Minutes per day')

# show legend
plt.legend()

# show graph
plt.show()
```

The data indicates that as an American grows older, they tend to spend more time being alone. The next data focuses around the percentage of American living alone by age.
